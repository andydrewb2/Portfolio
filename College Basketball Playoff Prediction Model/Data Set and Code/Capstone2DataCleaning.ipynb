```python
#After analyzing the data quite a bit the following values were determined to be able to fill in for null values in these columns
ncaa_games['neutral_site'] = ncaa_games['neutral_site'].fillna(False)
ncaa_games['a_flagrant_fouls'] = ncaa_games['a_flagrant_fouls'].fillna(0)
ncaa_games['a_player_tech_fouls'] = ncaa_games['a_player_tech_fouls'].fillna(0)
ncaa_games['a_team_tech_fouls'] = ncaa_games['a_team_tech_fouls'].fillna(0)
ncaa_games['a_coach_tech_fouls'] = ncaa_games['a_coach_tech_fouls'].fillna(0)
ncaa_games['conference_game'] = ncaa_games['a_coach_tech_fouls'].fillna(False)
ncaa_games['tournament'] = ncaa_games['tournament'].fillna('No')
ncaa_games['tournament_type'] = ncaa_games['tournament_type'].fillna('No')
ncaa_games['tournament_round'] = ncaa_games['tournament_round'].fillna('No')
ncaa_games['tournament_game_no'] = ncaa_games['tournament_game_no'].fillna('No')
ncaa_games['h_ejections'] = ncaa_games['h_ejections'].fillna(0)
ncaa_games['h_foulouts'] = ncaa_games['h_foulouts'].fillna(0)
ncaa_games['h_team_tech_fouls'] = ncaa_games['h_team_tech_fouls'].fillna(0)
ncaa_games['h_coach_tech_fouls'] = ncaa_games['h_coach_tech_fouls'].fillna(0)
ncaa_games['a_ejections'] = ncaa_games['a_ejections'].fillna(0)
ncaa_games['a_foulouts'] = ncaa_games['a_foulouts'].fillna(0)
print(ncaa_games.isna().sum())
```

```python
#Checking the different types of coverage values after seeing that there is a trend of missing values with rows that have boxscore in coverage
ncaa_games['coverage'].value_counts()
```

```python
#Boxscore does not contain enough values to do analysis, so dropping those rows is the best plan
#Additionally, rows that contain no values for conference alias and name alias seem to be wrong as well, so best plan is to drop those too
ncaa_games = ncaa_games[ncaa_games['coverage'] != 'boxscore']
ncaa_games = ncaa_games[ncaa_games['a_alias'].notnull()]
ncaa_games = ncaa_games[ncaa_games['a_conf_alias'].notnull()]
```

```python
print(ncaa_games.isna().sum().to_string())
```

```python
#Cleaning up columns that add no value to my project 
columns_to_drop = ['status', 'possession_arrow', 'venue_address','venue_zip','venue_country', 'h_logo_large', 'h_logo_medium', 'h_logo_small', 'a_logo_large', 'a_logo_medium', 'a_logo_small','created']
```
