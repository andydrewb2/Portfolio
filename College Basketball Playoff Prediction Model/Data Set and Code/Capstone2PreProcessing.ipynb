```python
#To handle date time, I will extract date and time to different columns, I will then make the date the difference between the first 
#date of the tournament and the date of this game, that will give an effect of distance from playoffs, which could have an effect
#on teams that perform better later in the season or have a devastating injury that causes them to be worse later 
#For time, I will make it the time, and then normalize is around the median time
```

```python
playoff_date = '2018-03-12'
playoff_date = pd.to_datetime(playoff_date)
playoff_date = playoff_date.date()  
type(playoff_date)
```

```python
method2_x['date'] = playoff_date - pd.to_datetime(method2_x['gametime']).dt.date 

method2_x['date'] = pd.to_timedelta(method2_x['date'])
method2_x['date'] = method2_x['date'].dt.days
print(method2_x['date'].head())
```
