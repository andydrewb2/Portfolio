```python
#Trying to determine if its worth tracking home and away scores/stats separately for each team and it looks like the average
#difference is 5 extra points at home than away in 2017, but when looking at the data as a whole, the numbers are all over the place
#and each team is significantly different.  Therefore, they probably do need to be tracked separately

import statistics

point_diff = []
for team in teams:
    temp = reg_season_2017[reg_season_2017['h_alias'] == team]
    h_points = temp['h_points_game'].median()
    temp2 = reg_season_2017[reg_season_2017['a_alias'] == team]
    a_points = temp['a_points_game'].median()
    point_diff.append(h_points - a_points)

print('mean')
print(statistics.mean(point_diff))
print('median')
print(statistics.median(point_diff))

print(point_diff)
```
